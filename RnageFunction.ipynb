# **Range Function** - Difference between the maximum and minimum values in a dataset

```python
# Using Python's built-in functions
scores = [78, 89, 65, 74, 80, 95, 88, 70, 82]
data_range = max(scores) - min(scores)
print(f"Range (using built-in functions): {data_range}")
```

```python
# Using NumPy
import numpy as np
scores = [78, 89, 65, 74, 80, 95, 88, 70, 82]
data_range_np = np.ptp(scores)  # PTP: Peak-to-Peak
print(f"Range (using NumPy): {data_range_np}")
```

```python
# Visualizing the Range using Matplotlib
import matplotlib.pyplot as plt

data = [4, 5, 6, 6, 7, 7, 7, 8, 8, 8, 8, 9, 9, 9, 10]

plt.hist(data, bins=6, edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title("Histogram of Data")
plt.axvline(x=min(data), color='red', linestyle='dashed', linewidth=1, label='Min')
plt.axvline(x=max(data), color='green', linestyle='dashed', linewidth=1, label='Max')
plt.legend()

# Save the histogram as an image
plt.savefig('assets/range_histogram.png', dpi=300)
plt.show()
```
