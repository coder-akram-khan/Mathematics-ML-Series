```python
import statistics as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample Data
scores = [85, 90, 92, 78, 88]
```

```python
# Calculate Mean & Standard Deviation
mean_scores = np.mean(scores)
std_scores = np.std(scores, ddof=1)  # ddof=1 for sample standard deviation
```

```python
# Plot Histogram
plt.hist(scores, bins=5, edgecolor='black', alpha=0.7, color='skyblue')
plt.axvline(mean_scores, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(mean_scores - std_scores, color='green', linestyle='dashed', linewidth=2, label='1 SD Left')
plt.axvline(mean_scores + std_scores, color='green', linestyle='dashed', linewidth=2, label='1 SD Right')

# Labels and Title
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Standard Deviation Visualization')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('assets/standard_deviation_histogram.png', dpi=300)
plt.show()
```
